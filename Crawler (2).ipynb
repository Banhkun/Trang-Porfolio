```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import json


class FacebookScraperCookies:
    def __init__(self, cookies_file):
        self.cookies_file = cookies_file
        self.driver = None

    def initialize_driver(self):
        options = Options()

        # MUST for Colab/Linux containers
        options.add_argument("--headless=new")
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")
        options.add_argument("--disable-gpu")
        options.add_argument("--remote-debugging-port=9222")

        # Extra stability
        options.add_argument("--window-size=1200,900")
        options.add_argument("--disable-notifications")
        options.add_argument("--disable-blink-features=AutomationControlled")

        options.add_argument(
            "--user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
        )

        self.driver = webdriver.Chrome(options=options)
        self.driver.execute_script(
            "Object.defineProperty(navigator, 'webdriver', {get: () => undefined})"
        )


    def load_cookies(self):
        """
        Load cookies into https://www.facebook.com (IMPORTANT!)
        """
        with open(self.cookies_file, "r", encoding="utf-8") as f:
            cookies = json.load(f)

        # MUST open correct domain first
        self.driver.get("https://www.facebook.com/")
        time.sleep(2)

        added = 0
        for cookie in cookies:
            c = cookie.copy()

            # Selenium needs "expiry" not "expirationDate"
            if "expirationDate" in c and "expiry" not in c:
                c["expiry"] = int(c["expirationDate"])

            # Remove problematic fields
            c.pop("hostOnly", None)
            c.pop("session", None)
            c.pop("storeId", None)
            c.pop("id", None)
            c.pop("sameSite", None)  # safest: remove, Chrome decides
            c.pop("expirationDate", None)

            # Make sure domain works
            # If cookie domain is ".facebook.com" it's OK for www
            # If it is "m.facebook.com", it may fail for www
            # We'll force it:
            if "domain" in c:
                c["domain"] = ".facebook.com"

            try:
                self.driver.add_cookie(c)
                added += 1
            except Exception as e:
                # Skip invalid cookies
                pass

        print(f"‚úÖ Added cookies: {added}")

        # Refresh and verify login
        self.driver.get("https://www.facebook.com/")
        time.sleep(3)

        if "login" in self.driver.current_url.lower():
            print("‚ùå Cookies loaded but still redirected to login.")
        else:
            print("‚úÖ Logged in using cookies!")

    def navigate_to_page(self, page_url):
        self.driver.get(page_url)
        WebDriverWait(self.driver, 15).until(
            EC.presence_of_element_located((By.TAG_NAME, "body"))
        )
        time.sleep(3)

    def slow_scroll(self, times=5):
        for i in range(times):
            self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2.5)
            print(f"Scroll {i+1}/{times}")

    def extract_posts_with_bs(self):
        soup = BeautifulSoup(self.driver.page_source, "html.parser")

        posts_data = []

        # FB changes a lot; we try multiple patterns
        post_blocks = soup.find_all("div", {"data-ad-preview": "message"})

        for block in post_blocks:
            try:
                post_text = block.get_text(" ", strip=True)
                if not post_text:
                    continue

                posts_data.append({
                    "post_text": post_text
                })
            except:
                continue

        # remove duplicates by post_text
        unique = {}
        for p in posts_data:
            unique[p["post_text"]] = p

        return list(unique.values())

    def scrape_posts(self, page_url, max_posts=20, scroll_rounds=10):
        self.navigate_to_page(page_url)

        all_posts = []
        seen = set()

        for i in range(scroll_rounds):
            posts = self.extract_posts_with_bs()

            for p in posts:
                t = p["post_text"]
                if t not in seen:
                    seen.add(t)
                    all_posts.append(p)

            print(f"‚úÖ Round {i+1}: total unique posts = {len(all_posts)}")

            if len(all_posts) >= max_posts:
                break

            self.slow_scroll(times=1)

        return all_posts[:max_posts]

    def print_posts(self, posts):
        for i, p in enumerate(posts, 1):
            print("=" * 60)
            print(f"Post {i}")
            print(p["post_text"][:800])

    def close(self):
        if self.driver:
            self.driver.quit()


if __name__ == "__main__":
    COOKIES_FILE = "/content/www.facebook.com_cookies.json"
    TARGET_PAGE = "https://www.facebook.com/groups/tuyendungmc"

    scraper = FacebookScraperCookies(COOKIES_FILE)

    try:
        scraper.initialize_driver()
        scraper.load_cookies()

        posts = scraper.scrape_posts(TARGET_PAGE, max_posts=10, scroll_rounds=8)
        scraper.print_posts(posts)

    finally:
        scraper.close()
```

```python
from playwright.async_api import async_playwright, Page, Browser
from bs4 import BeautifulSoup
import asyncio
import json
import re
import hashlib
from datetime import datetime
from difflib import SequenceMatcher
from typing import List, Dict, Optional, Set
import gspread
from google.auth import default
from telegram import Bot
from urllib.parse import urljoin, unquote


class ContentDeduplicator:
    """Handle content similarity detection to identify duplicate posts"""

    @staticmethod
    def normalize_text(text: str) -> str:
        """Normalize text for comparison"""
        text = re.sub(r'\s+', ' ', text.strip())
        text = re.sub(r'[^\w\s√Ä-·ªπ]', '', text, flags=re.UNICODE)
        return text.lower()

    @staticmethod
    def get_content_hash(text: str) -> str:
        """Generate hash for exact duplicate detection"""
        normalized = ContentDeduplicator.normalize_text(text)
        return hashlib.md5(normalized.encode('utf-8')).hexdigest()

    @staticmethod
    def similarity_ratio(text1: str, text2: str) -> float:
        """Calculate similarity between two texts (0-1)"""
        norm1 = ContentDeduplicator.normalize_text(text1)
        norm2 = ContentDeduplicator.normalize_text(text2)
        return SequenceMatcher(None, norm1, norm2).ratio()

    @staticmethod
    def is_duplicate(new_text: str, new_hash: str, existing_posts: List[Dict],
                    threshold: float = 0.85) -> Optional[Dict]:
        """Check if new_text is similar to any existing post"""
        for post in existing_posts:
            existing_hash = post.get('Content Hash') or post.get('content_hash', '')
            if existing_hash and existing_hash == new_hash:
                return post

            existing_content = post.get('Full Content') or post.get('content', '')
            if existing_content:
                similarity = ContentDeduplicator.similarity_ratio(new_text, existing_content)
                if similarity >= threshold:
                    return post

        return None


class GoogleSheetsStorage:
    """Manage post storage in Google Sheets"""

    def __init__(self, spreadsheet_name: str):
        self.spreadsheet_name = spreadsheet_name
        self.gc = None
        self.sheet = None
        self.worksheet = None
        self._initialize()

    def _initialize(self):
        """Initialize Google Sheets connection"""
        try:
            creds, _ = default()
            self.gc = gspread.authorize(creds)

            try:
                self.sheet = self.gc.open(self.spreadsheet_name)
                print(f"‚úÖ Opened existing spreadsheet: {self.spreadsheet_name}")
            except gspread.SpreadsheetNotFound:
                self.sheet = self.gc.create(self.spreadsheet_name)
                print(f"‚úÖ Created new spreadsheet: {self.spreadsheet_name}")

            try:
                self.worksheet = self.sheet.sheet1
            except:
                self.worksheet = self.sheet.get_worksheet(0)

            first_row = self.worksheet.row_values(1)
            if not first_row or first_row[0] != 'Timestamp':
                headers = [
                    'Timestamp', 'Group Name', 'Group URL', 'Post Link',
                    'Content Hash', 'Content Preview', 'Full Content',
                    'Sent to Telegram', 'First Seen', 'Last Seen'
                ]
                self.worksheet.update('A1:J1', [headers])
                print("‚úÖ Added headers to worksheet")
            else:
                print("‚úÖ Worksheet already has headers")

        except Exception as e:
            print(f"‚ùå Error initializing Google Sheets: {e}")
            raise

    def get_all_posts(self) -> List[Dict]:
        """Get all posts from the sheet"""
        try:
            records = self.worksheet.get_all_records()
            return records
        except Exception as e:
            print(f"‚ö†Ô∏è  Error fetching posts: {e}")
            return []

    def add_post(self, post_data: Dict) -> bool:
        """Add a new post to the sheet"""
        try:
            row = [
                post_data.get('timestamp', ''),
                post_data.get('group_name', ''),
                post_data.get('group_url', ''),
                post_data.get('post_link', ''),
                post_data.get('content_hash', ''),
                post_data.get('content_preview', ''),
                post_data.get('full_content', ''),
                post_data.get('sent_to_telegram', 'No'),
                post_data.get('first_seen', ''),
                post_data.get('last_seen', '')
            ]
            self.worksheet.append_row(row)
            return True
        except Exception as e:
            print(f"‚ùå Error adding post: {e}")
            return False

    def update_post_status(self, content_hash: str, sent: bool = True):
        """Update the telegram sent status of a post"""
        try:
            cell = self.worksheet.find(content_hash)
            if cell:
                row = cell.row
                self.worksheet.update_cell(row, 8, 'Yes' if sent else 'No')
                self.worksheet.update_cell(row, 10, datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
                return True
        except Exception as e:
            print(f"‚ö†Ô∏è  Error updating post status: {e}")
        return False

    def get_recent_posts(self, hours: int = 24) -> List[Dict]:
        """Get posts from the last N hours for comparison"""
        try:
            all_posts = self.get_all_posts()
            print(f"      üì• Retrieved {len(all_posts)} total posts from sheet")
            recent = all_posts[-500:] if len(all_posts) > 500 else all_posts
            print(f"      üîç Using {len(recent)} recent posts for duplicate checking")
            return recent
        except Exception as e:
            print(f"      ‚ö†Ô∏è  Error in get_recent_posts: {e}")
            return []


class FacebookGroupMonitor:
    """Monitor a single Facebook group using Playwright"""

    def __init__(self, group_name: str, group_url: str, page: Page):
        self.group_name = group_name
        self.group_url = group_url
        self.page = page

    def should_filter_post(self, content: str) -> tuple[bool, str]:
        """Check if post should be filtered out based on content rules"""
        content_lower = content.lower()

        has_hanoi = "h√† n·ªôi" in content_lower or "ha noi" in content_lower or "hanoi" in content_lower
        has_mc_nam = "mc nam" in content_lower
        has_nu = "n·ªØ" in content_lower or "nu" in content_lower or "n∆∞" in content_lower

        if has_hanoi or (has_mc_nam and not has_nu):
            return True, "Contains 'H√† N·ªôi' or 'MC Nam' without 'N·ªØ'"

        return False, ""

    async def navigate_to_group(self):
        """Navigate to the group page"""
        print(f"    üåê Navigating to {self.group_url}")
        try:
            # Use 'domcontentloaded' instead of 'networkidle' - much more reliable for Facebook
            await self.page.goto(self.group_url, wait_until='domcontentloaded', timeout=60000)
        except Exception as e:
            print(f"      ‚ö†Ô∏è Navigation warning: {e}")
            # Try one more time with 'load'
            try:
                await self.page.goto(self.group_url, wait_until='load', timeout=60000)
            except:
                pass

        # Wait for main content to be present
        try:
            await self.page.wait_for_selector('div[role="main"]', timeout=15000)
        except:
            # Fallback: just wait for any content
            await asyncio.sleep(5)

        # Additional wait for content to render
        await asyncio.sleep(3)
        print(f"    ‚úÖ Page loaded")

    async def scroll_feed(self, times: int = 5):
        """Scroll the feed to load more posts with smart waiting"""
        print(f"    üìú Scrolling feed {times} times...")

        for i in range(times):
            # Get current scroll height
            prev_height = await self.page.evaluate('document.body.scrollHeight')

            # Scroll to bottom
            await self.page.evaluate('window.scrollTo(0, document.body.scrollHeight)')

            # Wait for new content to load
            try:
                await self.page.wait_for_function(
                    f'document.body.scrollHeight > {prev_height}',
                    timeout=5000
                )
            except:
                # Timeout is fine, content might not load
                pass

            # Additional wait for rendering
            await asyncio.sleep(2)
            print(f"      Scroll {i+1}/{times}")

            # Check if we've hit the bottom
            is_at_bottom = await self.page.evaluate('''
                () => {
                    const scrollTop = window.pageYOffset || document.documentElement.scrollTop;
                    const scrollHeight = document.body.scrollHeight;
                    const clientHeight = window.innerHeight;
                    return scrollTop + clientHeight >= scrollHeight - 100;
                }
            ''')

            if is_at_bottom:
                current_height = await self.page.evaluate('document.body.scrollHeight')
                if current_height == prev_height:
                    print(f"      ‚ö†Ô∏è  Reached bottom early at scroll {i+1}")
                    break

    async def extract_posts(self) -> List[Dict]:
        """Extract posts from the current page with improved link detection"""

        # IMPORTANT: Wait extra time for all links to render
        print(f"    ‚è≥ Waiting for content to fully render...")
        await asyncio.sleep(3)  # Extra wait to ensure links are loaded

        content = await self.page.content()
        soup = BeautifulSoup(content, "html.parser")
        posts_data = []

        # Find all post containers - USE THE SAME SELECTOR THAT WORKS
        post_blocks = soup.find_all("div", {"data-ad-preview": "message"})

        if not post_blocks:
            # Fallback: try role="article"
            post_blocks = soup.find_all("div", {"role": "article"})
            print(f"    üîç Using fallback selector, found {len(post_blocks)} article blocks")

        print(f"    üîç Found {len(post_blocks)} post blocks")

        for idx, block in enumerate(post_blocks):
            try:
                post_text = block.get_text(" ", strip=True)
                if not post_text or len(post_text) < 10:
                    continue

                # Apply content filtering
                should_skip, reason = self.should_filter_post(post_text)
                if should_skip:
                    continue

                # Find post link with improved logic
                post_link = self._find_post_link(block)

                # Generate content hash
                content_hash = ContentDeduplicator.get_content_hash(post_text)

                posts_data.append({
                    "group_name": self.group_name,
                    "group_url": self.group_url,
                    "post_link": post_link,
                    "content": post_text,
                    "content_hash": content_hash,
                    "content_preview": post_text[:200],
                    "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                })

            except Exception as e:
                print(f"      ‚ö†Ô∏è  Error extracting post {idx}: {e}")
                continue

        # Remove duplicates within this batch
        unique = {}
        for p in posts_data:
            unique[p["content_hash"]] = p

        result = list(unique.values())

        # Count how many posts have links
        with_links = sum(1 for p in result if p['post_link'])
        print(f"    ‚úÖ Extracted {len(result)} unique posts ({with_links} with links, {len(result) - with_links} without)")

        return result


    def _find_post_link(self, block) -> Optional[str]:
        if not block:
            return None

        def clean_url(href: str) -> str:
            href = href.strip()

            # decode things like %2Fgroups%2F...
            href = unquote(href)

            # convert relative -> absolute
            href = urljoin("https://www.facebook.com", href)

            # remove FB tracking garbage but keep important ids
            href = href.split("&__cft__")[0]
            href = href.split("?__cft__")[0]
            href = href.split("&__tn__")[0]

            return href.rstrip("/")

        def is_post_permalink(href: str) -> bool:
            if not href:
                return False

            # reject profile/user links
            bad = [
                "/user/",
                "/people/",
                "/profile.php",
                "/members/",
                "/login",
                "/help",
                "/settings",
            ]
            if any(x in href for x in bad):
                return False

            # accept ONLY true post permalinks
            good = [
                "/posts/",
                "/permalink/",
                "story_fbid=",
                "pfbid",
            ]
            return any(x in href for x in good)

        # search around the block but keep it limited
        candidates = []

        # 1) inside the closest role=article container (BEST)
        article = block.find_parent("div", {"role": "article"})
        if article:
            candidates += article.find_all("a", href=True)

        # 2) fallback: inside parent
        if block.parent:
            candidates += block.parent.find_all("a", href=True)

        # 3) fallback: some nearby anchors
        candidates += block.find_all_next("a", href=True, limit=30)

        for a in candidates:
            href = a.get("href", "")
            if is_post_permalink(href):
                return clean_url(href)

        return None



class MultiGroupFacebookMonitor:
    """Monitor multiple Facebook groups simultaneously with Playwright"""

    def __init__(self, cookies_file: str, telegram_token: str, chat_id: str,
                 spreadsheet_name: str, similarity_threshold: float = 0.85):
        self.cookies_file = cookies_file
        self.telegram_token = telegram_token
        self.chat_id = chat_id
        self.playwright = None
        self.browser = None
        self.context = None
        self.page = None
        self.bot = Bot(token=telegram_token)
        self.storage = GoogleSheetsStorage(spreadsheet_name)
        self.deduplicator = ContentDeduplicator()
        self.similarity_threshold = similarity_threshold
        self.groups: List[FacebookGroupMonitor] = []

    def add_group(self, group_name: str, group_url: str):
        """Add a group to monitor"""
        # Will create monitor instance after browser is initialized
        self.groups.append({
            'name': group_name,
            'url': group_url
        })
        print(f"‚úÖ Added group: {group_name}")

    async def initialize_browser(self):
        """Initialize Playwright browser with stealth settings"""
        print("üöÄ Initializing Playwright browser...")

        self.playwright = await async_playwright().start()

        self.browser = await self.playwright.chromium.launch(
            headless=True,
            args=[
                '--no-sandbox',
                '--disable-setuid-sandbox',
                '--disable-dev-shm-usage',
                '--disable-blink-features=AutomationControlled',
                '--disable-gpu'
            ]
        )

        # Create context with realistic viewport and user agent
        self.context = await self.browser.new_context(
            viewport={'width': 1920, 'height': 1080},
            user_agent='Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
            locale='en-US',
            timezone_id='Asia/Singapore'
        )

        # Remove webdriver detection
        await self.context.add_init_script("""
            Object.defineProperty(navigator, 'webdriver', {
                get: () => undefined
            });
        """)

        self.page = await self.context.new_page()

        # Now create monitor instances with the page
        monitors = []
        for group in self.groups:
            monitor = FacebookGroupMonitor(group['name'], group['url'], self.page)
            monitors.append(monitor)
        self.groups = monitors

        print("‚úÖ Browser initialized")

    async def load_cookies(self) -> bool:
        """Load Facebook cookies into browser context"""
        print("üç™ Loading cookies...")

        with open(self.cookies_file, "r", encoding="utf-8") as f:
            cookies = json.load(f)

        # Navigate to Facebook first with relaxed wait condition
        try:
            await self.page.goto("https://www.facebook.com/", wait_until='domcontentloaded', timeout=30000)
        except:
            await self.page.goto("https://www.facebook.com/", wait_until='load', timeout=30000)

        await asyncio.sleep(2)

        # Convert and add cookies
        playwright_cookies = []
        for cookie in cookies:
            playwright_cookie = {
                'name': cookie['name'],
                'value': cookie['value'],
                'domain': cookie.get('domain', '.facebook.com'),
                'path': cookie.get('path', '/'),
            }

            # Add optional fields
            if 'expirationDate' in cookie:
                playwright_cookie['expires'] = int(cookie['expirationDate'])
            if 'httpOnly' in cookie:
                playwright_cookie['httpOnly'] = cookie['httpOnly']
            if 'secure' in cookie:
                playwright_cookie['secure'] = cookie['secure']

            playwright_cookies.append(playwright_cookie)

        await self.context.add_cookies(playwright_cookies)
        print(f"‚úÖ Added {len(playwright_cookies)} cookies")

        # Reload Facebook to apply cookies
        try:
            await self.page.goto("https://www.facebook.com/", wait_until='domcontentloaded', timeout=30000)
        except:
            await self.page.goto("https://www.facebook.com/", wait_until='load', timeout=30000)

        await asyncio.sleep(3)

        # Check if logged in
        current_url = self.page.url
        if "login" in current_url.lower():
            print("‚ùå Still redirected to login")
            return False
        else:
            print("‚úÖ Logged in successfully!")
            return True

    async def send_telegram_message(self, text: str) -> bool:
        """Send message to Telegram"""
        try:
            await self.bot.send_message(chat_id=self.chat_id, text=text, parse_mode='HTML')
            return True
        except Exception as e:
            print(f"    ‚ùå Telegram error: {e}")
            return False

    async def process_group(self, group: FacebookGroupMonitor, scroll_rounds: int = 5) -> int:
        """Process a single group and return count of new posts"""
        print(f"\n  üì± Processing: {group.group_name}")

        try:
            await group.navigate_to_group()
            await group.scroll_feed(times=scroll_rounds)

            print(f"    üîé Extracting posts...")
            posts = await group.extract_posts()

            # Get recent posts for comparison
            print(f"    üìä Loading existing posts from Google Sheets...")
            existing_posts = self.storage.get_recent_posts(hours=72)
            print(f"    üìä Loaded {len(existing_posts)} existing posts for comparison")

            new_posts_count = 0
            duplicate_count = 0

            for i, post in enumerate(posts, 1):
                print(f"    üîç Checking post {i}/{len(posts)}...")

                # Check for duplicates
                duplicate = ContentDeduplicator.is_duplicate(
                    post['content'],
                    post['content_hash'],
                    existing_posts,
                    threshold=self.similarity_threshold
                )

                if duplicate:
                    duplicate_count += 1
                    dup_group = duplicate.get('Group Name', 'unknown group')
                    dup_time = duplicate.get('First Seen', 'unknown time')
                    print(f"    ‚è≠Ô∏è  DUPLICATE: Similar to post from '{dup_group}' at {dup_time}")
                    continue

                # New unique post!
                print(f"    ‚ú® NEW POST FOUND!")
                print(f"       Hash: {post['content_hash'][:16]}...")
                print(f"       Link: {post['post_link'] or 'No link found'}")

                # Save to Google Sheets
                post_data = {
                    'timestamp': post['timestamp'],
                    'group_name': post['group_name'],
                    'group_url': post['group_url'],
                    'post_link': post['post_link'] or 'N/A',
                    'content_hash': post['content_hash'],
                    'content_preview': post['content_preview'],
                    'full_content': post['content'],
                    'sent_to_telegram': 'No',
                    'first_seen': post['timestamp'],
                    'last_seen': post['timestamp']
                }

                print(f"    üíæ Saving to Google Sheets...")
                if self.storage.add_post(post_data):
                    # Send to Telegram
                    message = self._format_telegram_message(post)
                    print(f"    üì§ Sending to Telegram...")
                    if await self.send_telegram_message(message):
                        self.storage.update_post_status(post['content_hash'], sent=True)
                        new_posts_count += 1
                        print(f"    ‚úÖ Successfully sent to Telegram")
                    else:
                        print(f"    ‚ö†Ô∏è  Failed to send to Telegram")

                    await asyncio.sleep(2)  # Rate limiting
                else:
                    print(f"    ‚ö†Ô∏è  Failed to save to Google Sheets")

                # Add to existing posts for this session
                existing_posts.append({
                    'Content Hash': post['content_hash'],
                    'Full Content': post['content'],
                    'Group Name': post['group_name'],
                    'First Seen': post['timestamp']
                })

            print(f"    üìä {group.group_name}: {new_posts_count} new, {duplicate_count} duplicates")
            return new_posts_count

        except Exception as e:
            print(f"    ‚ùå Error processing {group.group_name}: {e}")
            import traceback
            traceback.print_exc()
            return 0

    def _format_telegram_message(self, post: Dict) -> str:
        """Format post data for Telegram"""
        link_text = f'üîó <a href="{post["post_link"]}">View Post</a>' if post['post_link'] else 'üîó No direct link'

        content_preview = post['content'][:300]
        if len(post['content']) > 300:
            content_preview += "..."

        return f"""üîî <b>New Post</b>

üì± <b>Group:</b> {post['group_name']}

üìù <b>Content:</b>
{content_preview}

{link_text}

‚è∞ {post['timestamp']}"""

    async def monitor_all_groups(self, check_interval: int = 300, scroll_rounds: int = 5):
        """Monitor all groups in a loop"""
        print(f"\nüöÄ Starting multi-group monitor")
        print(f"üìä Monitoring {len(self.groups)} groups:")
        for g in self.groups:
            print(f"   ‚Ä¢ {g.group_name}")
        print(f"‚è±Ô∏è  Check interval: {check_interval}s ({check_interval/60:.1f} min)")
        print(f"üéØ Similarity threshold: {self.similarity_threshold}")

        iteration = 0
        while True:
            try:
                iteration += 1
                print(f"\n{'='*70}")
                print(f"üîÑ Check #{iteration} - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
                print(f"{'='*70}")

                total_new = 0
                for group in self.groups:
                    new_count = await self.process_group(group, scroll_rounds=scroll_rounds)
                    total_new += new_count

                print(f"\n‚úÖ Cycle complete: {total_new} total new posts across all groups")
                print(f"üí§ Waiting {check_interval}s until next check...")
                print(f"   Next check: {datetime.fromtimestamp(asyncio.get_event_loop().time() + check_interval).strftime('%Y-%m-%d %H:%M:%S')}")

                await asyncio.sleep(check_interval)

            except KeyboardInterrupt:
                raise
            except Exception as e:
                print(f"\n‚ùå Error in monitor loop: {e}")
                import traceback
                traceback.print_exc()

                print("‚è≥ Waiting 60s before retry...")
                await asyncio.sleep(60)

                # Try to recover
                try:
                    await self.close()
                    await self.initialize_browser()
                    if not await self.load_cookies():
                        print("‚ùå Failed to reload cookies!")
                except Exception as e2:
                    print(f"‚ùå Recovery failed: {e2}")

    async def close(self):
        """Clean up resources"""
        if self.browser:
            await self.browser.close()
        if self.playwright:
            await self.playwright.stop()


async def main():
    """Main entry point"""

    # ========== CONFIGURATION ==========
    COOKIES_FILE = "/content/www.facebook.com_cookies.json"

    # For Google Colab - get credentials from userdata
    from google.colab import userdata
    TELEGRAM_BOT_TOKEN = userdata.get("TELEGRAM_BOT_TOKEN")
    CHAT_ID = userdata.get("CHAT_ID")

    SPREADSHEET_NAME = "MC post spreadsheet"

    # Groups to monitor
    GROUPS = [
        {
            "name": "Tuy·ªÉn D·ª•ng MC",
            "url": "https://www.facebook.com/groups/tuyendungmc"
        },
        {
            "name": "MC TI·ªÜC C∆Ø·ªöI/ S·ª∞ KI·ªÜN CHUY√äN NGHI·ªÜP & K·∫æT N·ªêI JOB",
            "url": "https://www.facebook.com/groups/236474123644327"
        },
        {
            "name": "MC GROUP - C·ªông ƒë·ªìng MC Chuy√™n nghi·ªáp To√†n Qu·ªëc",
            "url": "https://www.facebook.com/groups/mcvanminh"
        },
    ]

    CHECK_INTERVAL = 180  # 3 minutes
    SCROLL_ROUNDS = 8
    SIMILARITY_THRESHOLD = 0.85  # 85% similarity = duplicate
    # ===================================

    monitor = MultiGroupFacebookMonitor(
        cookies_file=COOKIES_FILE,
        telegram_token=TELEGRAM_BOT_TOKEN,
        chat_id=CHAT_ID,
        spreadsheet_name=SPREADSHEET_NAME,
        similarity_threshold=SIMILARITY_THRESHOLD
    )

    try:
        # Add all groups
        for group in GROUPS:
            monitor.add_group(group["name"], group["url"])

        print("\nüîß Initializing browser...")
        await monitor.initialize_browser()

        print("üç™ Loading cookies...")
        if not await monitor.load_cookies():
            print("‚ùå Failed to login!")
            return

        # Start monitoring
        await monitor.monitor_all_groups(
            check_interval=CHECK_INTERVAL,
            scroll_rounds=SCROLL_ROUNDS
        )

    except KeyboardInterrupt:
        print("\n‚ö†Ô∏è  Stopped by user")
    except Exception as e:
        print(f"‚ùå Fatal error: {e}")
        import traceback
        traceback.print_exc()
    finally:
        await monitor.close()


if __name__ == "__main__":
    # For Jupyter/Colab compatibility
    try:
        import nest_asyncio
        nest_asyncio.apply()
    except ImportError:
        pass

    asyncio.run(main())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import json
import asyncio
from telegram import Bot
from datetime import datetime
import pickle
import os

class FacebookTelegramMonitor:
    def __init__(self, cookies_file, telegram_token, chat_id):
        self.cookies_file = cookies_file
        self.telegram_token = telegram_token
        self.chat_id = chat_id
        self.driver = None
        self.bot = Bot(token=telegram_token)
        self.sent_posts_file = "sent_posts.pkl"
        self.sent_posts = self.load_sent_posts()

    def load_sent_posts(self):
        """Load previously sent post IDs"""
        if os.path.exists(self.sent_posts_file):
            with open(self.sent_posts_file, 'rb') as f:
                return pickle.load(f)
        return set()

    def save_sent_posts(self):
        """Save sent post IDs"""
        with open(self.sent_posts_file, 'wb') as f:
            pickle.dump(self.sent_posts, f)

    def initialize_driver(self):
        options = Options()
        options.add_argument("--headless=new")
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")
        options.add_argument("--disable-gpu")
        options.add_argument("--remote-debugging-port=9222")
        options.add_argument("--window-size=1200,900")
        options.add_argument("--disable-notifications")
        options.add_argument("--disable-blink-features=AutomationControlled")
        options.add_argument(
            "--user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
        )

        self.driver = webdriver.Chrome(options=options)
        self.driver.execute_script(
            "Object.defineProperty(navigator, 'webdriver', {get: () => undefined})"
        )

    def load_cookies(self):
        """Load cookies into https://www.facebook.com"""
        with open(self.cookies_file, "r", encoding="utf-8") as f:
            cookies = json.load(f)

        self.driver.get("https://www.facebook.com/")
        time.sleep(2)

        added = 0
        for cookie in cookies:
            c = cookie.copy()
            if "expirationDate" in c and "expiry" not in c:
                c["expiry"] = int(c["expirationDate"])

            c.pop("hostOnly", None)
            c.pop("session", None)
            c.pop("storeId", None)
            c.pop("id", None)
            c.pop("sameSite", None)
            c.pop("expirationDate", None)

            if "domain" in c:
                c["domain"] = ".facebook.com"

            try:
                self.driver.add_cookie(c)
                added += 1
            except:
                pass

        print(f"‚úÖ Added cookies: {added}")

        self.driver.get("https://www.facebook.com/")
        time.sleep(3)

        if "login" in self.driver.current_url.lower():
            print("‚ùå Cookies loaded but still redirected to login.")
            return False
        else:
            print("‚úÖ Logged in using cookies!")
            return True

    def navigate_to_page(self, page_url):
        self.driver.get(page_url)
        WebDriverWait(self.driver, 15).until(
            EC.presence_of_element_located((By.TAG_NAME, "body"))
        )
        time.sleep(3)

    def slow_scroll(self, times=5):
        for i in range(times):
            self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2.5)
            print(f"  Scroll {i+1}/{times}")

    def extract_posts_with_links(self):
        """Extract posts with their links - USING YOUR WORKING SELECTOR"""
        soup = BeautifulSoup(self.driver.page_source, "html.parser")
        posts_data = []

        # USE THE SAME SELECTOR THAT WORKS IN YOUR OLD CODE
        post_blocks = soup.find_all("div", {"data-ad-preview": "message"})

        print(f"  Found {len(post_blocks)} post blocks with data-ad-preview=message")

        for block in post_blocks:
            try:
                post_text = block.get_text(" ", strip=True)
                if not post_text:
                    continue

                # Now find the link - look in parent elements
                post_link = None

                # Method 1: Search in parent for links
                parent = block.parent
                if parent:
                    # Look for timestamp links (usually has post URL)
                    links = parent.find_all("a", href=True)
                    for a in links:
                        href = a.get("href", "")
                        # Facebook post links contain these patterns
                        if any(pattern in href for pattern in ["/posts/", "/permalink/", "story_fbid=", "/groups/", "/photo/"]):
                            post_link = href
                            break

                # Method 2: If not found, search the whole block tree
                if not post_link:
                    links = block.find_all_next("a", href=True, limit=20)
                    for a in links:
                        href = a.get("href", "")
                        if any(pattern in href for pattern in ["/posts/", "/permalink/", "story_fbid="]):
                            post_link = href
                            break

                # Method 3: Search previous siblings
                if not post_link:
                    prev_siblings = block.find_all_previous("a", href=True, limit=20)
                    for a in prev_siblings:
                        href = a.get("href", "")
                        if any(pattern in href for pattern in ["/posts/", "/permalink/", "story_fbid="]):
                            post_link = href
                            break

                # If still no link, create a unique ID from text
                if not post_link:
                    # Use text hash as ID
                    post_id = f"text_{hash(post_text[:200])}"
                    post_link = None
                else:
                    # Make full URL
                    if post_link.startswith("/"):
                        post_link = "https://www.facebook.com" + post_link
                    elif not post_link.startswith("http"):
                        post_link = "https://www.facebook.com/" + post_link

                    # Clean URL and use as ID
                    post_id = post_link.split("?")[0]

                posts_data.append({
                    "post_id": post_id,
                    "post_link": post_link,
                    "post_text": post_text[:500],  # First 500 chars
                    "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                })

            except Exception as e:
                print(f"  Error extracting post: {e}")
                continue

        # Remove duplicates by post_id
        unique = {}
        for p in posts_data:
            unique[p["post_id"]] = p

        result = list(unique.values())
        print(f"  Extracted {len(result)} unique posts")
        return result

    async def send_telegram_message(self, text):
        """Send message to Telegram"""
        try:
            await self.bot.send_message(chat_id=self.chat_id, text=text, parse_mode='HTML')
            print(f"  ‚úÖ Sent to Telegram")
            return True
        except Exception as e:
            print(f"  ‚ùå Error sending to Telegram: {e}")
            return False

    async def check_and_send_new_posts(self, page_url, scroll_rounds=5):
        """Check for new posts and send to Telegram"""
        print(f"\nüîç Navigating to page...")
        self.navigate_to_page(page_url)

        # Scroll to load posts (using your working method)
        print(f"üìú Scrolling {scroll_rounds} times...")
        self.slow_scroll(times=scroll_rounds)

        print(f"üîé Extracting posts...")
        posts = self.extract_posts_with_links()

        new_posts_count = 0

        for i, post in enumerate(posts, 1):
            post_id = post["post_id"]

            if post_id not in self.sent_posts:
                print(f"\n  üìÆ New post #{i}/{len(posts)}")

                # Format message
                if post['post_link']:
                    message = f"""üîî <b>New Post in Facebook Group</b>

üìù {post['post_text']}

üîó <a href="{post['post_link']}">View Post</a>

‚è∞ {post['timestamp']}"""
                else:
                    message = f"""üîî <b>New Post in Facebook Group</b>

üìù {post['post_text']}

‚è∞ {post['timestamp']}

<i>(No direct link found)</i>"""

                # Send to Telegram
                success = await self.send_telegram_message(message)

                if success:
                    self.sent_posts.add(post_id)
                    new_posts_count += 1
                    self.save_sent_posts()

                # Don't spam - wait between messages
                await asyncio.sleep(2)
            else:
                print(f"  ‚è≠Ô∏è  Post #{i}/{len(posts)} - Already sent (skipping)")

        print(f"\nüìä Summary: Found {len(posts)} total posts, sent {new_posts_count} new posts")
        return new_posts_count

    async def monitor_loop(self, page_url, check_interval=300, scroll_rounds=5):
        """
        Continuously monitor the page
        check_interval: seconds between checks (default 5 minutes)
        scroll_rounds: how many times to scroll on each check
        """
        print(f"üöÄ Starting monitor for: {page_url}")
        print(f"‚è±Ô∏è  Check interval: {check_interval} seconds ({check_interval/60:.1f} minutes)")
        print(f"üìú Scroll rounds per check: {scroll_rounds}")

        iteration = 0
        while True:
            try:
                iteration += 1
                print(f"\n{'='*70}")
                print(f"üîÑ Check #{iteration} - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
                print(f"{'='*70}")

                await self.check_and_send_new_posts(page_url, scroll_rounds=scroll_rounds)

                print(f"\nüí§ Waiting {check_interval} seconds ({check_interval/60:.1f} minutes) until next check...")
                print(f"   Next check at: {datetime.fromtimestamp(time.time() + check_interval).strftime('%Y-%m-%d %H:%M:%S')}")
                await asyncio.sleep(check_interval)

            except Exception as e:
                print(f"\n‚ùå Error in monitor loop: {e}")
                import traceback
                traceback.print_exc()

                print("‚è≥ Waiting 60 seconds before retry...")
                await asyncio.sleep(60)

                # Try to reinitialize driver if needed
                try:
                    print("üîÑ Attempting to reinitialize driver...")
                    if self.driver:
                        self.driver.quit()
                    self.initialize_driver()
                    if not self.load_cookies():
                        print("‚ùå Failed to reload cookies!")
                except Exception as e2:
                    print(f"‚ùå Failed to reinitialize: {e2}")

    def close(self):
        if self.driver:
            self.driver.quit()


async def main():
    # Configuration
    COOKIES_FILE = "/content/www.facebook.com_cookies.json"
    TELEGRAM_BOT_TOKEN = userdata.get("TELEGRAM_BOT_TOKEN")
    CHAT_ID = userdata.get("CHAT_ID")
    TARGET_PAGE = "https://www.facebook.com/groups/tuyendungmc"
    CHECK_INTERVAL = 180  # Check every 3 minutes (180 seconds)
    SCROLL_ROUNDS = 5  # How many times to scroll on each check

    monitor = FacebookTelegramMonitor(COOKIES_FILE, TELEGRAM_BOT_TOKEN, CHAT_ID)

    try:
        print("üîß Initializing driver...")
        monitor.initialize_driver()

        print("üç™ Loading cookies...")
        if not monitor.load_cookies():
            print("‚ùå Failed to login with cookies!")
            return

        # Start monitoring loop
        await monitor.monitor_loop(
            TARGET_PAGE,
            check_interval=CHECK_INTERVAL,
            scroll_rounds=SCROLL_ROUNDS
        )

    except KeyboardInterrupt:
        print("\n‚ö†Ô∏è  Stopped by user")
    except Exception as e:
        print(f"‚ùå Fatal error: {e}")
        import traceback
        traceback.print_exc()
    finally:
        monitor.close()


if __name__ == "__main__":
    # For Google Colab
    import nest_asyncio
    nest_asyncio.apply()

    asyncio.run(main())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import json
import asyncio
from telegram import Bot
from datetime import datetime
import hashlib
from difflib import SequenceMatcher
import gspread
from google.auth import default
from typing import List, Dict, Optional
import re

class ContentDeduplicator:
    """Handle content similarity detection to identify duplicate posts"""

    @staticmethod
    def normalize_text(text: str) -> str:
        """Normalize text for comparison"""
        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text.strip())
        # Remove special characters but keep Vietnamese
        text = re.sub(r'[^\w\s√Ä-·ªπ]', '', text, flags=re.UNICODE)
        # Convert to lowercase
        return text.lower()

    @staticmethod
    def get_content_hash(text: str) -> str:
        """Generate hash for exact duplicate detection"""
        normalized = ContentDeduplicator.normalize_text(text)
        return hashlib.md5(normalized.encode('utf-8')).hexdigest()

    @staticmethod
    def similarity_ratio(text1: str, text2: str) -> float:
        """Calculate similarity between two texts (0-1)"""
        norm1 = ContentDeduplicator.normalize_text(text1)
        norm2 = ContentDeduplicator.normalize_text(text2)
        return SequenceMatcher(None, norm1, norm2).ratio()

    @staticmethod
    def is_duplicate(new_text: str, new_hash: str, existing_posts: List[Dict], threshold: float = 0.85) -> Optional[Dict]:
        """
        Check if new_text is similar to any existing post
        Returns the duplicate post if found, None otherwise
        """
        for post in existing_posts:
            # Exact match via hash - check the 'Content Hash' field
            existing_hash = post.get('Content Hash') or post.get('content_hash', '')
            if existing_hash and existing_hash == new_hash:
                return post

            # Similarity match - check both 'Full Content' and 'content' fields
            existing_content = post.get('Full Content') or post.get('content', '')
            if existing_content:
                similarity = ContentDeduplicator.similarity_ratio(new_text, existing_content)
                if similarity >= threshold:
                    return post

        return None


class GoogleSheetsStorage:
    """Manage post storage in Google Sheets using Colab auth"""

    def __init__(self, spreadsheet_name: str):
        self.spreadsheet_name = spreadsheet_name
        self.gc = None
        self.sheet = None
        self.worksheet = None
        self._initialize()

    def _initialize(self):
        """Initialize Google Sheets connection using Colab authentication"""
        try:
            # Authorize with Colab credentials
            creds, _ = default()
            self.gc = gspread.authorize(creds)

            # Try to open existing spreadsheet
            try:
                self.sheet = self.gc.open(self.spreadsheet_name)
                print(f"‚úÖ Opened existing spreadsheet: {self.spreadsheet_name}")
            except gspread.SpreadsheetNotFound:
                # Create new spreadsheet
                self.sheet = self.gc.create(self.spreadsheet_name)
                print(f"‚úÖ Created new spreadsheet: {self.spreadsheet_name}")

            # Get or create the first worksheet
            try:
                self.worksheet = self.sheet.sheet1
            except:
                self.worksheet = self.sheet.get_worksheet(0)

            # Check if headers exist
            first_row = self.worksheet.row_values(1)
            if not first_row or first_row[0] != 'Timestamp':
                # Add headers
                headers = [
                    'Timestamp', 'Group Name', 'Group URL', 'Post Link',
                    'Content Hash', 'Content Preview', 'Full Content',
                    'Sent to Telegram', 'First Seen', 'Last Seen'
                ]
                self.worksheet.update('A1:J1', [headers])
                print("‚úÖ Added headers to worksheet")
            else:
                print("‚úÖ Worksheet already has headers")

        except Exception as e:
            print(f"‚ùå Error initializing Google Sheets: {e}")
            raise

    def get_all_posts(self) -> List[Dict]:
        """Get all posts from the sheet"""
        try:
            records = self.worksheet.get_all_records()
            return records
        except Exception as e:
            print(f"‚ö†Ô∏è  Error fetching posts: {e}")
            return []

    def add_post(self, post_data: Dict) -> bool:
        """Add a new post to the sheet"""
        try:
            row = [
                post_data.get('timestamp', ''),
                post_data.get('group_name', ''),
                post_data.get('group_url', ''),
                post_data.get('post_link', ''),
                post_data.get('content_hash', ''),
                post_data.get('content_preview', ''),
                post_data.get('full_content', ''),
                post_data.get('sent_to_telegram', 'No'),
                post_data.get('first_seen', ''),
                post_data.get('last_seen', '')
            ]
            self.worksheet.append_row(row)
            return True
        except Exception as e:
            print(f"‚ùå Error adding post: {e}")
            return False

    def update_post_status(self, content_hash: str, sent: bool = True):
        """Update the telegram sent status of a post"""
        try:
            # Find the cell with this content hash
            cell = self.worksheet.find(content_hash)
            if cell:
                row = cell.row
                # Update "Sent to Telegram" column (H = 8) and "Last Seen" column (J = 10)
                self.worksheet.update_cell(row, 8, 'Yes' if sent else 'No')
                self.worksheet.update_cell(row, 10, datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
                return True
        except Exception as e:
            print(f"‚ö†Ô∏è  Error updating post status: {e}")
        return False

    def get_recent_posts(self, hours: int = 24) -> List[Dict]:
        """Get posts from the last N hours for comparison"""
        try:
            all_posts = self.get_all_posts()
            print(f"      üì• Retrieved {len(all_posts)} total posts from sheet")

            # Return last 500 posts for efficiency (covers ~48-72 hours typically)
            recent = all_posts[-500:] if len(all_posts) > 500 else all_posts
            print(f"      üîç Using {len(recent)} recent posts for duplicate checking")
            return recent
        except Exception as e:
            print(f"      ‚ö†Ô∏è  Error in get_recent_posts: {e}")
            return []


class FacebookGroupMonitor:
    """Monitor a single Facebook group"""

    def __init__(self, group_name: str, group_url: str, driver):
        self.group_name = group_name
        self.group_url = group_url
        self.driver = driver

    def should_filter_post(self, content: str) -> tuple[bool, str]:
        """
        Check if post should be filtered out based on content rules
        Returns: (should_filter, reason)
        """
        content_lower = content.lower()

        # Rule 1: If contains "H√† N·ªôi" or "MC Nam" but NOT "N·ªØ"/"Nu"
        has_hanoi = "h√† n·ªôi" in content_lower or "ha noi" in content_lower or "hanoi" in content_lower
        has_mc_nam = "mc nam" in content_lower
        has_nu = "n·ªØ" in content_lower or "nu" in content_lower or "n∆∞" in content_lower

        if has_hanoi or (has_mc_nam and not has_nu):
            return True, "Contains 'H√† N·ªôi' or 'MC Nam' without 'N·ªØ'"

        return False, ""

    def navigate_to_group(self):
        """Navigate to the group page"""
        self.driver.get(self.group_url)
        WebDriverWait(self.driver, 15).until(
            EC.presence_of_element_located((By.TAG_NAME, "body"))
        )
        time.sleep(3)

    def scroll_feed(self, times: int = 5):
        """Scroll the feed to load more posts"""
        for i in range(times):
            self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2.5)
            print(f"    Scroll {i+1}/{times}")

    def extract_posts(self) -> List[Dict]:
        """Extract posts from the current page"""
        soup = BeautifulSoup(self.driver.page_source, "html.parser")
        posts_data = []

        post_blocks = soup.find_all("div", {"data-ad-preview": "message"})
        print(f"    Found {len(post_blocks)} post blocks")

        for block in post_blocks:
            try:
                post_text = block.get_text(" ", strip=True)
                if not post_text or len(post_text) < 10:
                    continue
                should_skip, reason = self.should_filter_post(post_text)
                if should_skip:
                    print(f"  FILTERED: {reason}")  # uncomment for debug
                    continue
                # Find post link
                post_link = self._find_post_link(block)

                # Generate content hash
                content_hash = ContentDeduplicator.get_content_hash(post_text)

                posts_data.append({
                    "group_name": self.group_name,
                    "group_url": self.group_url,
                    "post_link": post_link,
                    "content": post_text,
                    "content_hash": content_hash,
                    "content_preview": post_text[:200],
                    "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                })

            except Exception as e:
                print(f"    Error extracting post: {e}")
                continue

        # Remove duplicates within this batch
        unique = {}
        for p in posts_data:
            unique[p["content_hash"]] = p

        result = list(unique.values())
        print(f"    Extracted {len(result)} unique posts from {self.group_name}")
        return result
    def _find_post_link(self, block) -> Optional[str]:
        if not block:
            return None

        def clean_url(href):
            if href.startswith("/"):
                href = "https://www.facebook.com" + href
            href = href.split("?", 1)[0].rstrip("/")
            return href

        def is_good_post_link(href):
            if not href:
                return False
            h = href.lower()
            # Strong patterns - must have one of these
            if any(x in h for x in ["/posts/", "/permalink/", "story_fbid="]):
                return True
            # /groups/... but must look like actual post (contains numbers at end or /posts/)
            if "/groups/" in h and re.search(r'/\d{10,}$', h):
                return True
            return False

        # 1. Best chance: inside role="article"
        article = block.find_parent(attrs={"role": "article"})
        if article:
            for a in article.find_all("a", href=True):
                href = a.get("href", "")
                if is_good_post_link(href):
                    return clean_url(href)

        # 2. Look for timestamp-looking link first (very reliable when present)
        time_keywords = ["ph√∫t", "gi·ªù", "ng√†y", "tu·∫ßn", "th√°ng", "hr", "min", "hour", "day", "week", "month", "ago"]
        for a in block.find_all("a", href=True):
            href = a.get("href", "")
            aria = (a.get("aria-label") or "").lower()
            if is_good_post_link(href) and any(kw in aria for kw in time_keywords):
                return clean_url(href)

        # 3. Any good link in parent chain (up to 6 levels)
        current = block
        for _ in range(6):
            if not current:
                break
            for a in current.find_all("a", href=True):
                href = a.get("href", "")
                if is_good_post_link(href):
                    return clean_url(href)
            current = current.parent

        return None

class MultiGroupFacebookMonitor:
    """Monitor multiple Facebook groups simultaneously"""

    def __init__(self, cookies_file: str, telegram_token: str, chat_id: str,
                 spreadsheet_name: str, similarity_threshold: float = 0.85):
        self.cookies_file = cookies_file
        self.telegram_token = telegram_token
        self.chat_id = chat_id
        self.driver = None
        self.bot = Bot(token=telegram_token)
        self.storage = GoogleSheetsStorage(spreadsheet_name)
        self.deduplicator = ContentDeduplicator()
        self.similarity_threshold = similarity_threshold
        self.groups: List[FacebookGroupMonitor] = []

    def add_group(self, group_name: str, group_url: str):
        """Add a group to monitor"""
        monitor = FacebookGroupMonitor(group_name, group_url, self.driver)
        self.groups.append(monitor)
        print(f"‚úÖ Added group: {group_name}")

    def initialize_driver(self):
        """Initialize Selenium WebDriver"""
        options = Options()
        options.add_argument("--headless=new")
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")
        options.add_argument("--disable-gpu")
        options.add_argument("--remote-debugging-port=9222")
        options.add_argument("--window-size=1200,900")
        options.add_argument("--disable-notifications")
        options.add_argument("--disable-blink-features=AutomationControlled")
        options.add_argument(
            "--user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
        )

        self.driver = webdriver.Chrome(options=options)
        self.driver.execute_script(
            "Object.defineProperty(navigator, 'webdriver', {get: () => undefined})"
        )

        # Update driver reference for all groups
        for group in self.groups:
            group.driver = self.driver

    def load_cookies(self) -> bool:
        """Load Facebook cookies"""
        with open(self.cookies_file, "r", encoding="utf-8") as f:
            cookies = json.load(f)

        self.driver.get("https://www.facebook.com/")
        time.sleep(2)

        added = 0
        for cookie in cookies:
            c = cookie.copy()
            if "expirationDate" in c and "expiry" not in c:
                c["expiry"] = int(c["expirationDate"])

            c.pop("hostOnly", None)
            c.pop("session", None)
            c.pop("storeId", None)
            c.pop("id", None)
            c.pop("sameSite", None)
            c.pop("expirationDate", None)

            if "domain" in c:
                c["domain"] = ".facebook.com"

            try:
                self.driver.add_cookie(c)
                added += 1
            except:
                pass

        print(f"‚úÖ Added cookies: {added}")

        self.driver.get("https://www.facebook.com/")
        time.sleep(3)

        if "login" in self.driver.current_url.lower():
            print("‚ùå Still redirected to login")
            return False
        else:
            print("‚úÖ Logged in successfully!")
            return True

    async def send_telegram_message(self, text: str) -> bool:
        """Send message to Telegram"""
        try:
            await self.bot.send_message(chat_id=self.chat_id, text=text, parse_mode='HTML')
            return True
        except Exception as e:
            print(f"    ‚ùå Telegram error: {e}")
            return False

    async def process_group(self, group: FacebookGroupMonitor, scroll_rounds: int = 5) -> int:
        """Process a single group and return count of new posts"""
        print(f"\n  üì± Processing: {group.group_name}")

        try:
            group.navigate_to_group()
            print(f"    üìú Scrolling...")
            group.scroll_feed(times=scroll_rounds)

            print(f"    üîé Extracting posts...")
            posts = group.extract_posts()

            # Get recent posts for comparison
            print(f"    üìä Loading existing posts from Google Sheets...")
            existing_posts = self.storage.get_recent_posts(hours=72)
            print(f"    üìä Loaded {len(existing_posts)} existing posts for comparison")

            new_posts_count = 0
            duplicate_count = 0

            for i, post in enumerate(posts, 1):
                print(f"    üîç Checking post {i}/{len(posts)}...")

                # Check if this content already exists
                duplicate = ContentDeduplicator.is_duplicate(
                    post['content'],
                    post['content_hash'],
                    existing_posts,
                    threshold=self.similarity_threshold
                )

                if duplicate:
                    duplicate_count += 1
                    dup_group = duplicate.get('Group Name', 'unknown group')
                    dup_time = duplicate.get('First Seen', 'unknown time')
                    print(f"    ‚è≠Ô∏è  DUPLICATE: Similar to post from '{dup_group}' at {dup_time}")
                    print(f"       Hash: {post['content_hash'][:16]}... (match: {duplicate.get('Content Hash', '')[:16]}...)")
                    continue

                # New unique post!
                print(f"    ‚ú® NEW POST FOUND!")
                print(f"       Hash: {post['content_hash'][:16]}...")
                print(f"       Preview: {post['content_preview'][:80]}...")

                # Save to Google Sheets
                post_data = {
                    'timestamp': post['timestamp'],
                    'group_name': post['group_name'],
                    'group_url': post['group_url'],
                    'post_link': post['post_link'] or 'N/A',
                    'content_hash': post['content_hash'],
                    'content_preview': post['content_preview'],
                    'full_content': post['content'],
                    'sent_to_telegram': 'No',
                    'first_seen': post['timestamp'],
                    'last_seen': post['timestamp']
                }

                print(f"    üíæ Saving to Google Sheets...")
                if self.storage.add_post(post_data):
                    # Send to Telegram
                    message = self._format_telegram_message(post)
                    print(f"    üì§ Sending to Telegram...")
                    if await self.send_telegram_message(message):
                        self.storage.update_post_status(post['content_hash'], sent=True)
                        new_posts_count += 1
                        print(f"    ‚úÖ Successfully sent to Telegram")
                    else:
                        print(f"    ‚ö†Ô∏è  Failed to send to Telegram")

                    await asyncio.sleep(2)  # Rate limiting
                else:
                    print(f"    ‚ö†Ô∏è  Failed to save to Google Sheets")

                # Add to existing posts for this session to prevent duplicates within same run
                existing_posts.append({
                    'Content Hash': post['content_hash'],
                    'Full Content': post['content'],
                    'Group Name': post['group_name'],
                    'First Seen': post['timestamp']
                })

            print(f"    üìä {group.group_name}: {new_posts_count} new, {duplicate_count} duplicates")
            return new_posts_count

        except Exception as e:
            print(f"    ‚ùå Error processing {group.group_name}: {e}")
            import traceback
            traceback.print_exc()
            return 0

    def _format_telegram_message(self, post: Dict) -> str:
        """Format post data for Telegram"""
        link_text = f'üîó <a href="{post["post_link"]}">View Post</a>' if post['post_link'] else 'üîó No direct link'

        content_preview = post['content'][:300]
        if len(post['content']) > 300:
            content_preview += "..."

        return f"""üîî <b>New Post</b>

üì± <b>Group:</b> {post['group_name']}

üìù <b>Content:</b>
{content_preview}

{link_text}

‚è∞ {post['timestamp']}"""

    async def monitor_all_groups(self, check_interval: int = 300, scroll_rounds: int = 5):
        """Monitor all groups in a loop"""
        print(f"\nüöÄ Starting multi-group monitor")
        print(f"üìä Monitoring {len(self.groups)} groups:")
        for g in self.groups:
            print(f"   ‚Ä¢ {g.group_name}")
        print(f"‚è±Ô∏è  Check interval: {check_interval}s ({check_interval/60:.1f} min)")
        print(f"üéØ Similarity threshold: {self.similarity_threshold}")

        iteration = 0
        while True:
            try:
                iteration += 1
                print(f"\n{'='*70}")
                print(f"üîÑ Check #{iteration} - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
                print(f"{'='*70}")

                total_new = 0
                for group in self.groups:
                    new_count = await self.process_group(group, scroll_rounds=scroll_rounds)
                    total_new += new_count

                print(f"\n‚úÖ Cycle complete: {total_new} total new posts across all groups")
                print(f"üí§ Waiting {check_interval}s until next check...")
                print(f"   Next check: {datetime.fromtimestamp(time.time() + check_interval).strftime('%Y-%m-%d %H:%M:%S')}")

                await asyncio.sleep(check_interval)

            except KeyboardInterrupt:
                raise
            except Exception as e:
                print(f"\n‚ùå Error in monitor loop: {e}")
                import traceback
                traceback.print_exc()

                print("‚è≥ Waiting 60s before retry...")
                await asyncio.sleep(60)

                # Try to recover
                try:
                    if self.driver:
                        self.driver.quit()
                    self.initialize_driver()
                    if not self.load_cookies():
                        print("‚ùå Failed to reload cookies!")
                except Exception as e2:
                    print(f"‚ùå Recovery failed: {e2}")

    def close(self):
        """Clean up resources"""
        if self.driver:
            self.driver.quit()


async def main():
    """Main entry point"""

    # ========== CONFIGURATION ==========
    COOKIES_FILE = "/content/www.facebook.com_cookies.json"
    TELEGRAM_BOT_TOKEN = userdata.get("TELEGRAM_BOT_TOKEN")
    CHAT_ID = userdata.get("CHAT_ID")
    SPREADSHEET_NAME = "MC post spreadsheet"

    # Groups to monitor
    GROUPS = [
        {
            "name": "Tuy·ªÉn D·ª•ng MC",
            "url": "https://www.facebook.com/groups/tuyendungmc"
        },
        {
            "name": "MC TI·ªÜC C∆Ø·ªöI/ S·ª∞ KI·ªÜN CHUY√äN NGHI·ªÜP & K·∫æT N·ªêI JOB",
            "url": "https://www.facebook.com/groups/236474123644327"
        },
        {
            "name": "MC GROUP - C·ªông ƒë·ªìng MC Chuy√™n nghi·ªáp To√†n Qu·ªëc",
            "url": "https://www.facebook.com/groups/mcvanminh"
        },
        # Add more groups here
    ]

    CHECK_INTERVAL = 180  # 3 minutes
    SCROLL_ROUNDS = 5
    SIMILARITY_THRESHOLD = 0.85  # 85% similarity = duplicate
    # ===================================

    monitor = MultiGroupFacebookMonitor(
        cookies_file=COOKIES_FILE,
        telegram_token=TELEGRAM_BOT_TOKEN,
        chat_id=CHAT_ID,
        spreadsheet_name=SPREADSHEET_NAME,
        similarity_threshold=SIMILARITY_THRESHOLD
    )

    try:
        # Add all groups
        for group in GROUPS:
            monitor.add_group(group["name"], group["url"])

        print("\nüîß Initializing driver...")
        monitor.initialize_driver()

        print("üç™ Loading cookies...")
        if not monitor.load_cookies():
            print("‚ùå Failed to login!")
            return

        # Start monitoring
        await monitor.monitor_all_groups(
            check_interval=CHECK_INTERVAL,
            scroll_rounds=SCROLL_ROUNDS
        )

    except KeyboardInterrupt:
        print("\n‚ö†Ô∏è  Stopped by user")
    except Exception as e:
        print(f"‚ùå Fatal error: {e}")
        import traceback
        traceback.print_exc()
    finally:
        monitor.close()


if __name__ == "__main__":
    import nest_asyncio
    nest_asyncio.apply()

    asyncio.run(main())
```

```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configuration
keyword = 'covid'
num_pages = 3

# Set up Chrome driver
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=chrome_options)


# Search for keyword and navigate to first search result page
driver.get('https://kenh14.vn/')
# locate the search input element and enter the search keyword
search_input = driver.find_element_by_css_selector("p[data-placeholder='T√¨m ki·∫øm...']")
search_input.send_keys("covid")

# locate the search submit button and click it
search_submit = driver.find_element_by_css_selector(".t-search-submit-btn")
search_submit.click()

# wait for the search results to load and print the page title
wait = WebDriverWait(driver, 10)
search_results = wait.until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, "h3.knswli-title>a")))
print(f"Found {len(search_results)} search results for 'covid' on kenh14.vn\n")
print(driver.title)

# loop through each search result and scrape article content and comments
for i in range(len(search_results)):
    # re-locate the search result elements on each iteration
    search_results = wait.until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, "h3.knswli-title>a")))
    result = search_results[i]

    result_link = result.get_attribute("href")
    result_title = result.text
    print(f"\n{result_title}\n{result_link}\n")

    # navigate to the search result page
    driver.get(result_link)

    # scrape the article content
    article_content = driver.find_element_by_css_selector("div.knc-content")
    print(article_content.text)
# quit the driver
driver.quit()
```

#Scrape Google Images


```python
import os
import time
import urllib.request

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from IPython.display import Image, display

# Configuration
keyword = 'puppies'
image_folder = 'images'
num_images = 50

# Create the image folder if it does not exist
if not os.path.exists(image_folder):
    os.makedirs(image_folder)

# Set up Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=chrome_options)

# Navigate to Google Images and search for the keyword
driver.get('https://www.google.com/imghp')
search_box = driver.find_element_by_name('q')
search_box.send_keys(keyword)
search_box.send_keys(Keys.RETURN)

# Scroll down to load more images
last_height = driver.execute_script('return document.body.scrollHeight')

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(2)
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == last_height:
        break
    last_height = new_height

# Download images and display in Colab
images = driver.find_elements_by_css_selector("img.rg_i")
count = 0

for image in images:
    try:
        image_url = image.get_attribute('src')
        if image_url is not None:
            file_name = f"{count}.jpg"
            urllib.request.urlretrieve(image_url, f"{image_folder}/{file_name}")
            count += 1
            if count >= num_images:
                break
            display(Image(filename=f"{image_folder}/{file_name}"))
    except Exception as e:
        print(f"Error: {str(e)}")
        continue

print(f"Downloaded {count} images to the '{image_folder}' folder.")
driver.quit()
```
